```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
!conda install -c anaconda xlrd --yes
```

```python
df_canada = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
                       sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2
                      )

print('Data downloaded and read into a dataframe!')
```

```python
df_canada.head()
```

```python
print(df_canada.shape)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1, inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_canada.head()
```

```python
df_canada.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)

df_canada.columns=list(map(str, df_canada.columns))
df_canada.set_index('Country', inplace=True)
df_canada['Total']=df_canada.sum(axis=1)
years=list(map(str,range(1984,2014)))
print('data dimensions:', df_canada.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
print('matplotlib_Version:', mpl.__version__)
```

```python
df_continents=df_canada.groupby('Continent',axis=0).sum()
df_continents.head()
```

```python
df_tot= pd.DataFrame(df_canada[years].sum(axis=0))


df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.columns= ['year', 'Total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='Total', figsize=(10,6), color='darkblue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_tot['year']
y = df_tot['Total']
fit = np.polyfit(x, y , deg=1)
fit
```

```python
df_tot.plot(kind = 'scatter', x='year', y='Total', figsize=(10,6), color='darkblue')
plt.title('Total number of immigrants of canada from 1984-2013')
plt.xlabel(' Year')
plt.ylabel('number of immigrants')
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

```python
df_canada_t=df_canada[years].transpose()
df_canada_t.head()
```

```python
# cast the Years (the index) to type int
df_canada_t.index = map(int, df_canada_t.index)

# let's label the index. This will automatically be the column name when we reset the index
df_canada_t.index.name = 'year'

# reset index to bring the Year in as a column
df_canada_t.reset_index(inplace=True)

# view the changes
df_canada_t.head()
```

```python
# normalize Brazil data
norm_brazil = (df_canada_t['Brazil'] - df_canada_t['Brazil'].min()) / (df_canada_t['Brazil'].max() - df_canada_t['Brazil'].min())

# normalize Argentina data
norm_argentina = (df_canada_t['Argentina'] - df_canada_t['Argentina'].min()) / (df_canada_t['Argentina'].max() - df_canada_t['Argentina'].min())
```

```python
# Brazil
ax0 = df_canada_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,                  # transparency
                    color='green',
                    s=norm_brazil * 2000 + 10,  # pass in weights 
                    xlim=(1975, 2015)
                   )

# Argentina
ax1 = df_canada_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina*2000+10,
                    ax = ax0
                   )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

```python
norm_china=(df_canada_t['China']-df_canada_t['China'].min())/(df_canada_t['China'].max()-df_canada_t['China'].min())
norm_india=(df_canada_t['India']-df_canada_t['India'].min())/(df_canada_t['India'].max()-df_canada_t['India'].min())
```

```python
ax0=df_canada_t.plot(kind='scatter',
                     x='Year',
                     y='China',
                     figsize=(14,8),
                     alpha=0.5,
                     color='green',
                     s=norm_china*2000+10,
                     xlim=(1975,2015)
                     )

ax1=df_canada_t.plot(kind='scatter',
                     x='Year',
                     y='India',
                     figsize=(14,8),
                     alpha=0.5,
                     color='red',
                     s=norm_india*2000+10,
                     ax=ax0)
ax0.set_title('immigration from china and india btw 1983 to 2013')
ax0.set_ylabel('Number of immigrants')
ax0.legend(['China', 'India'], loc='upper left', fontsize='x-large')
```
